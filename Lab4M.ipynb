```python
#Exercise 3
reset()
```

```python
t, C1, C2 = var('t, C1, C2')
x = function('x')(t)
y = function('y')(t)
deq1 = diff(x, t) == x + t
deq2 = diff(y, t) == -2 * x + 4 * t

sols = desolve_system([deq1, deq2], [x, y], [0, C1, C2])
sol_x(t, C1, C2) = sols[0].rhs()
sol_y(t, C1, C2) = sols[1].rhs()
print(sol_x, sol_y)
```

```python
#a
sol_x_1 = sol_x(t, 3, 0)
sol_y_1 = sol_y(t, 3, 0)
print('Solutions for first ivp: ')
print(sol_x_1)
print(sol_y_1)

sol_x_2 = sol_x(t, 0, 3)
sol_y_2 = sol_y(t, 0, 3)

sol_x_3 = sol_x(t, -3, 0)
sol_y_3 = sol_y(t, -3, 0)

sol_x_4 = sol_x(t, 0, -3)
sol_y_4 = sol_y(t, 0, -3)
```

```python
#b
def printLimit(sol_x, sol_y):
    print(limit(sol_x, t=infinity))
    print(limit(sol_y, t=infinity))
    print()

printLimit(sol_x_1, sol_y_1)
printLimit(sol_x_2, sol_y_2)
printLimit(sol_x_3, sol_y_3)
printLimit(sol_x_4, sol_y_4)
```

```python
#c
f1(u, v) = u + v
f2(u, v) = - 2 * u + 4 * v
n = sqrt(f1(u, v) ^ 2 + f2(u, v) ^ 2)
g =plot_vector_field([f1(u, v) / n, f2(u, v) / n], [u, -10, 10], [v, -10, 10], aspect_ratio=1)

g = g + parametric_plot((sol_x_1(t), sol_y_1(t)), (t, -10, 10), color='blue')
g = g + parametric_plot((sol_x_2(t), sol_y_2(t)), (t, -10, 10), color='blue')
g = g + parametric_plot((sol_x_3(t), sol_y_3(t)), (t, -10, 10), color='blue')
g = g + parametric_plot((sol_x_4(t), sol_y_4(t)), (t, -10, 10), color='blue')
g.show(xmin=-5, xmax=5, ymin=-5, ymax=5)
```
