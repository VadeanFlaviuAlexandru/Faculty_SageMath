```python
t = var('t')
x(t) = (1 - cos(t)) * cos(t)
y(t) = (1 - cos(t)) * sin(t)
parametric_plot((x(t), y(t)), (t, 0, 2*pi))
```

```python
x(t) = sin(3*t) * cos(t)
y(t) = sin(3*t) * sin(t)
parametric_plot((x(t), y(t)), (t, 0, 2 * pi))
```

```python
x(t) = t - sin(t)
y(t) = 1 - cos(t)
parametric_plot((x(t), y(t)), (t, 0, 6 * pi))
```

```python
s = var('s')
f(t, s) = 1 - (s * cos(4*t) * cos(t)) / (sqrt(1 - s^2 * cos(4*t)^2 * sin(t) ^ 2))

g = parametric_plot((f(t - pi / 2, 0.1), f(t, 0.1)), (t, 0, 2 * pi))
for i in range(2, 11):
    g = g + parametric_plot((f(t - pi / 2, i / 10), f(t, i / 10)), (t, 0, 2 * pi))
    
show(g)
```
