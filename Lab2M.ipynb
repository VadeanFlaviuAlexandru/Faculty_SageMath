```python
x = var('x')
y = function('y')(x)
eqd1 = diff(y, x) == 2 * x * (1 + y ^ 2)
desolve(eqd1, y)
```

```python
eqd2 = (x^2 - 1) * diff(y, x) + 2 * x * y^2 == 0
desolve(eqd2, y)
```

```python
eqd3 = diff(y, x, 2) + y == sin(x) + cos(x)
desolve(eqd3, y)
```

```python
eqd4 = diff(y, x, 2) - y == e ^ (2*x)
desolve(eqd4, y)
```

```python
eqd5 = diff(y, x, 2) - diff(y, x) == 1 / (1 + e^x)
desolve(eqd5, y)
```

```python
eqd6 = diff(y, x) == 1 + y^2
sol6 = desolve(eqd6, y, ics=[0, 1])
sol6
```

```python
x, k, _C = var('x, k, _C')
y = function('y')(x)
deq = diff(y, x) + (k / x) * y == x ^ 3
general_solution(x, k, _C) = desolve(deq, [y, x])
general_solution(x, k, _C)
```

```python
list_sol = [general_solution(x, 1, i) for i in range(-2, 2)]
list_sol
plot(list_sol, x, -10, 10, ymin=-50, ymax=50)
```

```python
ivp_solution(x, k) = desolve(deq, [y, x], ics=[1, 0])
ivp_solution
```

```python
sols = [plot(ivp_solution(x, k/20), x, -2, 2) for k in [0..5]]
animate(sols)
```

```python
x, a, b, _C = var('x, a, b, _C')
y = function('y')(x)
deq = diff(y, x) == y(x) + a * x + b
ivp_solution(x, a, b) = desolve(deq, [y, x], ics=[0, 1])
ivp_solution
```

```python
syst = [ivp_solution(1, a, b) == -5 + 4*e^2, ivp_solution(2, a, b) == -7 + 4*e^2]
solve(syst, a, b)
```
